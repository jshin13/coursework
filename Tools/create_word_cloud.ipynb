```python
import os
```

```python
folder_path = 'D:\GDrive\CloudData\Photo-Exchange-Hub\DAIS_Vids'
for d, _, f in os.walk(folder_path):
    print(len(f))
```

```python
file_names = []
for f_i in f:
    file_names.append(os.path.basename(f_i).split('.')[0])
file_names
```

```python
from wordcloud import WordCloud
```

```python
comb_names = ''
for i in file_names:
    comb_names = comb_names + ''.join(i)

print(comb_names[:10])
```

```python
wc = WordCloud(collocations=False, background_color='white').generate(comb_names)
```

```python
wc.to_image()
```

```python
data = wc.to_svg(embed_font=True)
with open('wordcloud_dais.svg', 'w+') as f:
    f.write(data)
```
